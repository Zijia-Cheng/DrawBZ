```python
from BZdrawer import BZ
import numpy as np
import matplotlib.pyplot as plt
```

```python
#defining the reciprocal lattice vectors
kvector = np.array([[1.56, 0.00, -0.618],
                    [0.00, 1.56, -0.618],
                    [0.00, 0.00, 1.236]])
```

```python
#initializing the class
BZ1 = BZ(kvector)
```

```python
#drawing the bulk BZ
BZ1.bulkBZ()
fig, ax = BZ1.draw_bulkBZ()
#remove all the ticks and labels
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])


plt.show()
```

```python
dis = 2
direc = np.array([0.0, 0.0, 1.0])
BZ1.surfaceBZ(dis, direc)
```
